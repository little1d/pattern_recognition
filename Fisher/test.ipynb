```python
from utils import fisher_discriminant_analysis
import numpy as np

def cal_cov_and_avg(samples):
    mu = np.mean(samples, axis=0)
    cov_m = np.zeros((samples.shape[1], samples.shape[1]))
    for i in range(samples.shape[1]):
        s = samples.iloc[i,:]
        t = (s - mu).to_numpy()
        cov_m += t*t.T
    return cov_m, mu


def fisher_discriminant_analysis(X, y):
    index1 = np.array(y[y['class'] == 0].index) # 获取类别为'0'的 index
    index2 = np.array(y[y['class'] == 1].index) # 获取类别为'1'的 index
    con_1, u1 = cal_cov_and_avg(X.iloc[index1])
    con_2, u2 = cal_cov_and_avg(X.iloc[index2])
    S_W = con_1 + con_2
    u, s, v = np.linalg.svd(S_W)
    # 投影方向向量wß
    S_W_inv = np.dot(np.dot(v.T, np.linalg.inv(np.diag(s))), u.T)
    W = np.dot(S_W_inv,u1 - u2)
    return W

w = fisher_discriminant_analysis(X, y)
```

```python
# 未测试

np.set_printoptions(precision=4)
def fisher_discriminant_analysis1(X,y):
    y_c = np.unique(y)
    mean_vector = [] # 类别的平均值
    for i in y_c:
        mean_vector.append(np.mean(X[y==i], axis=0))
        print('Mean  Vector class %s:%s\n' % (i, mean_vector[i]))
        # 类内
        S_W = np.zeros((X.shape[1], X.shape[1])) 
        for i in y_c:
            Xi = X[y==i] - mean_vector[i]
            S_W += np.asmatrix(Xi).T * np.asmatrix(Xi)
        print('Within-class scatter matrix: %s\n' % S_W)
        # 类间
        S_B = np.zeros((X.shape[1], X.shape[1]))
        mu = np.mean(X, axis=0) # 所有样本平均值
        for i in y_c:
            Ni = len(X[y==i])
            S_B += Ni * np.asmatrix(mean_vector[i] - mu).T * np.asmatrix(mean_vector[i] - mu)
        print('Between-class scatter matrix: %s\n' % S_B)
        # 计算变换矩阵的广义特征值
        eigvals, eigvecs = np.linalg.eig(np.linalg.inv(S_W) * S_B)  # 求特征值，特征向量
        np.testing.assert_array_almost_equal(np.asmatrix(np.linalg.inv(S_W) * S_B) * np.asmatrix(eigvecs[:, 0].reshape(4, 1)),
                                        eigvals[0] * np.asmatrix(eigvecs[:, 0].reshape(4, 1)), decimal=6, err_msg='',
                                        verbose=True)
        # sorting the eigenvectors by decreasing eigenvalues
        eig_pairs = [(np.abs(eigvals[i]), eigvecs[:, i]) for i in range(len(eigvals))]
        eig_pairs = sorted(eig_pairs, key=lambda k: k[0], reverse=True)
        W = np.hstack((eig_pairs[0][1].reshape(4, 1), eig_pairs[1][1].reshape(4, 1)))
        return W
    
W = fisher_discriminant_analysis1(X,y)
X_trans = X.dot(W)
assert X_trans.shape == (150, 2)
```
